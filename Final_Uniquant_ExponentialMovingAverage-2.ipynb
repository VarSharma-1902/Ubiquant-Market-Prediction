<b> The Parameters for Exponential Moving Average, include the Time Period as well alpha that depends upon the Chosen Time Period. </b><br><b> Alpha is also known as the Smoothing Factor.  </b>

```python
# Model parameters
time_period = 30
alpha = 2/(time_period+1)
```

```python
data['target']
```

```python
exp_moving_avg = data['target'].ewm(span=time_period, adjust=False).mean()
exp_moving_avg
```

```python
fig1 = plt.figure(figsize=(10,8))
ax1 = fig1.add_subplot(111,ylabel='target')
data['target'].plot(ax=ax1, color='b', label="Actual")
exp_moving_avg.plot(ax=ax1, color='r', label="Predicted")
plt.xlim(0,100)
plt.legend()
plt.show()
```

```python
from sklearn import metrics
print('Mean Absolute Error (MAE): ', metrics.mean_absolute_error(new_df['target'],exp_moving_avg))
print('Mean Squared Error (MSE): ', metrics.mean_squared_error(new_df['target'],exp_moving_avg))
print('Root Mean Squared Error (RMSE): ', np.sqrt(metrics.mean_squared_error(new_df['target'],exp_moving_avg)))
```
